## Gets the mean and std of the dataset

```python
import os
import torch
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
```

```python
# Constents
PATH ='./data/wood_dataset/Train'
BATCH_SIZE = 32
TREE_LABELS = {
    0: 'common beech',
    1: 'common walnut',
    2: 'chestnut',
    3: 'austrian oak',
    4: 'common alder',
    5: 'manna ash',
    6: 'european spruce',
    7: 'ailanthus',
    8: 'varnish tree',
    9: 'black locust',
    10: 'mediterranean cypress',
    11: 'sycamore'
}
```

```python
training_transforms = transforms.Compose([transforms.ToTensor()])
```

```python
train_dataset = torchvision.datasets.ImageFolder(root = PATH, transform=training_transforms)
```

```python
train_loader = torch.utils.data.DataLoader(dataset = train_dataset, batch_size=BATCH_SIZE, shuffle = True)
```

```python
examplars=iter(train_loader)
fig, axes=plt.subplots(4,4)
fig.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1, wspace=0.4, hspace=0.4)
for i in range(4):
    samples, labels=next(examplars)
    for j in range(4):
        axes[i][j].set_title(TREE_LABELS[labels[j].item()])
        axes[i][j].imshow(samples[j][0])
        axes[i][j].set_xticks([])
        axes[i][j].set_yticks([])
plt.show()
```

```python
def get_mean_std():
    mean = 0
    std = 0
    total_images = 0
    for images, _ in train_loader:
        current_batch_size = images.size(0)
        images_resized = images.view(current_batch_size, images.size(1), -1)
        mean += images_resized.mean(2).sum(0)
        std += images_resized.std(2).sum(0)
        total_images += current_batch_size
    mean /= total_images
    std /= total_images
    return mean, std
```

```python
mean, std = get_mean_std()
```

```python
print(mean)
print(std)
```
